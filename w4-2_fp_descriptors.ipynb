```python
# smiles 코드가 단일 구조가 아니라면 '.'이 포함되어 있음.
# 단일 구조가 아닌 물질들만 리스트를 확인하고, 제거해야 될 것과 제거하면 안되는 물질 구분해보기.
# 현재 대부분이 유기 화합물이기 때문에, 유기화합물이 염(salt)가 추가된 경우는, 염을 제거한 smiles 코드를 남겨야함.
# 무기화합물이라면 제거. 유효 성분이 몇개인지 확인이 어려운 경우도 제거.
```

```python
# 정제된 최종 SMILES를 담을 빈 리스트
cleaned_smiles_list = []

# 전체 데이터(df)를 대상으로 반복문을 실행합니다.
for index, row in df.iterrows():
    smiles = str(row['SMILES'])
    
    # [케이스 1] 처음부터 마침표('.')가 없는 '완벽한 단일 분자'인 경우
    if '.' not in smiles:
        cleaned_smiles_list.append(smiles) # 그대로 리스트에 통과!
        continue
        
    # --------------------------------------------------------
    # [케이스 2] 마침표('.')가 있는 혼합물인 경우 (이전 코드의 정제 과정 적용)
    
    # 유효 성분이 불명확한 경우 (더미 원자 '*' 포함) 제거
    if '*' in smiles:
        cleaned_smiles_list.append(None)
        continue
        
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        cleaned_smiles_list.append(None)
        continue
        
    # GetMolFrags를 사용하여 혼합물을 개별 조각으로 쪼개기
    frags = Chem.GetMolFrags(mol, asMols=True)
    largest_frag = None
    max_atoms = 0
    
    # 가장 원자가 많은 큰 조각(메인 유기물 뼈대) 찾기
    for frag in frags:
        num_atoms = frag.GetNumAtoms()
        if num_atoms > max_atoms:
            max_atoms = num_atoms
            largest_frag = frag
            
    # 선택된 가장 큰 조각이 무기화합물인지 확인 (탄소 'C' 포함 여부)
    has_carbon = any(atom.GetSymbol() == 'C' for atom in largest_frag.GetAtoms())
    
    if not has_carbon:
        # 탄소가 없으면 무기물이므로 제거
        cleaned_smiles_list.append(None)
    else:
        # 유기물이면 염이 떨어져 나간 깔끔한 상태로 저장
        cleaned_smiles = Chem.MolToSmiles(largest_frag)
        cleaned_smiles_list.append(cleaned_smiles)

# 원본 데이터프레임에 정제된 결과를 새로운 열로 추가합니다.
df['Cleaned_SMILES'] = cleaned_smiles_list

# 정제가 불가능해 None 처리된 데이터(무기물, 불명확 구조 등)를 최종 삭제합니다.
df_final = df.dropna(subset=['Cleaned_SMILES']).reset_index(drop=True)

# 최종 남은 데이터 개수 확인
print("최종적으로 분석에 사용할 데이터 개수:", len(df_final), "개")
```

```python
df_final
#smiles(다른 시작점이지만 같은 구조일 수 있음)->mol->smiles로 표준화 후 smiles 중복 제거
```

```python
# --- 위에서 작성한 정제 코드(df_final 생성) 이후에 바로 이어서 실행합니다 ---

standardized_smiles_list = []

# 1. 살아남은 Cleaned_SMILES를 순회하며 표준화(Canonicalization) 진행
for index, row in df_final.iterrows():
    smiles = row['Cleaned_SMILES']
    
    # SMILES -> Mol 객체로 변환
    mol = Chem.MolFromSmiles(smiles)
    
    if mol is not None:
        # Mol -> SMILES 재변환
        # (이 과정에서 RDKit이 자동으로 구조를 인식하여 동일한 규칙의 문자열로 통일해 줍니다)
        canonical_smiles = Chem.MolToSmiles(mol)
        standardized_smiles_list.append(canonical_smiles)
    else:
        standardized_smiles_list.append(None)

# 2. 표준화된 SMILES 결과를 데이터프레임에 새로운 열로 저장
df_final['Standardized_SMILES'] = standardized_smiles_list

# 변환 과정에서 문제가 생긴 데이터가 혹시 있다면 제거
df_final = df_final.dropna(subset=['Standardized_SMILES'])

# 3. 'Standardized_SMILES' 열을 기준으로 중복되는 분자를 제거 (Pandas 기능 활용)
# keep='first' 옵션을 통해 중복된 구조가 발견되면 맨 처음 나온 데이터 하나만 남깁니다.
df_deduplicated = df_final.drop_duplicates(subset=['Standardized_SMILES'], keep='first').reset_index(drop=True)

# 4. 중복 제거까지 마친 진짜 최종 데이터 개수 확인
print("표준화 및 중복 제거 완료! 최종 남은 데이터 개수:", len(df_deduplicated), "개")
```

```python
#https://www.rdkit.org/docs/GettingStartedInPython.html#list-of-available-descriptors
#fingerprint 종류는 매우 다양, 종류별로 모두 계산해서 저장해보기
```

```python
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs
from rdkit.Chem import MACCSkeys 

# 1. 4가지 기본 Fingerprint 생성기 준비
morgan_gen = AllChem.GetMorganGenerator(radius=2)
rdkit_gen = AllChem.GetRDKitFPGenerator()
atompair_gen = AllChem.GetAtomPairGenerator()
torsion_gen = AllChem.GetTopologicalTorsionGenerator()

valid_rows = []
morgan_fps, rdkit_fps, atompair_fps, torsion_fps = [], [], [], []
maccs_fps, pattern_fps = [], []

# 지문을 Numpy Array로 변환해주는 반복 사용 함수 (질문자님 코드 응용)
def fp_to_numpy(fp):
    fp_array = np.zeros((0,), dtype=int)
    DataStructs.ConvertToNumpyArray(fp, fp_array)
    return fp_array

# 2. 반복문 실행
for index, row in df.iterrows():
    smiles = str(row['SMILES'])
    
    # 단일 분자만 통과
    if '.' not in smiles:
        mol = Chem.MolFromSmiles(smiles)
        
        if mol is not None:
            # 생성된 지문을 numpy array로 변환하여 리스트에 추가
            morgan_fps.append(fp_to_numpy(morgan_gen.GetFingerprint(mol)))
            rdkit_fps.append(fp_to_numpy(rdkit_gen.GetFingerprint(mol)))
            atompair_fps.append(fp_to_numpy(atompair_gen.GetFingerprint(mol)))
            torsion_fps.append(fp_to_numpy(torsion_gen.GetFingerprint(mol)))
            
            maccs_fps.append(fp_to_numpy(MACCSkeys.GenMACCSKeys(mol)))
            pattern_fps.append(fp_to_numpy(Chem.PatternFingerprint(mol)))
            
            valid_rows.append(row)

# 3. 수집된 Numpy 배열 리스트들을 데이터프레임으로 변환
df_morgan = pd.DataFrame(morgan_fps)
df_rdkit = pd.DataFrame(rdkit_fps)
df_atompair = pd.DataFrame(atompair_fps)
df_torsion = pd.DataFrame(torsion_fps)
df_maccs = pd.DataFrame(maccs_fps)
df_pattern = pd.DataFrame(pattern_fps)

# 4. 각 지문의 열(Column) 이름에 접두사 붙이기
df_morgan.columns = [f'Morgan_{i}' for i in range(df_morgan.shape[1])]
df_rdkit.columns = [f'RDKit_{i}' for i in range(df_rdkit.shape[1])]
df_atompair.columns = [f'AtomPair_{i}' for i in range(df_atompair.shape[1])]
df_torsion.columns = [f'Torsion_{i}' for i in range(df_torsion.shape[1])]
df_maccs.columns = [f'MACCS_{i}' for i in range(df_maccs.shape[1])]
df_pattern.columns = [f'Pattern_{i}' for i in range(df_pattern.shape[1])]

# 원본 데이터도 표로 변환 후 인덱스 초기화
valid_df = pd.DataFrame(valid_rows).reset_index(drop=True)

# 5. 원본 데이터와 각각의 Numpy 변환 지문 데이터를 따로 합치고, 개별 파일로 저장
pd.concat([valid_df, df_morgan], axis=1).to_csv("morgan_fingerprints.csv", index=False)
pd.concat([valid_df, df_rdkit], axis=1).to_csv("rdkit_fingerprints.csv", index=False)
pd.concat([valid_df, df_atompair], axis=1).to_csv("atompair_fingerprints.csv", index=False)
pd.concat([valid_df, df_torsion], axis=1).to_csv("torsion_fingerprints.csv", index=False)
pd.concat([valid_df, df_maccs], axis=1).to_csv("maccs_fingerprints.csv", index=False)
pd.concat([valid_df, df_pattern], axis=1).to_csv("pattern_fingerprints.csv", index=False)

print("Numpy 배열 방식을 적용하여 6가지 Fingerprint가 개별 파일로 저장되었습니다!")
```
