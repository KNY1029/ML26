```python
import pandas as pd
df = pd.read_excel('skin_irritation.xlsx', sheet_name='Data_invivo')
df
```

```python
df['Assay'].value_counts()
```

```python
subset = df[df['Mixture']=='Chemical']
subset = subset[subset['Species']=='Human']
subset = subset[subset['Assay']=='Four-hour Human Patch Test']
subset
```

```python
subset['Endpoint'].value_counts()
```

```python
import matplotlib.pyplot as plt

# 1. 조건에 맞는 데이터 필터링
positive_df = subset[subset['Endpoint'] == 'Positive reaction']

# 1. 문자로 된 데이터를 강제로 숫자로 변환 
# (errors='coerce'를 넣으면 숫자로 바꿀 수 없는 이상한 텍스트는 빈칸(NaN)으로 처리해 줍니다)
positive_df['Response'] = pd.to_numeric(positive_df['Response'])

# 2. 숫자로 변환되었으므로 히스토그램 시각화 가능
positive_df['Response'].plot.hist()

# 3. 그래프 화면에 출력
plt.show()
```
